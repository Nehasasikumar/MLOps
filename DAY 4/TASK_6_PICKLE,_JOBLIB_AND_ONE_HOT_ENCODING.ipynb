```python
import pandas as pd
import pickle
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/multimedia_knowledge_dataset.csv")

# Input and Output
X = df[['source_credibility']]
y = df['trust_score']

# Train model
model = LinearRegression()
model.fit(X, y)

# Save model
with open("linear_model.pkl", "wb") as file:
    pickle.dump(model, file)

print("Model saved successfully")
```

```python
import pickle

# Load saved model
with open("linear_model.pkl", "rb") as file:
    loaded_model = pickle.load(file)

# Prediction
prediction = loaded_model.predict([[9]])

print("Predicted Trust Score:")
print(prediction[0])
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from joblib import dump

# Load dataset
df = pd.read_csv("/content/multimedia_knowledge_dataset.csv")

# Input and output
X = df[['source_credibility']]
y = df['trust_score']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Save model
dump(model, "linear_model.joblib")

print("Model saved successfully")
```

```python
from joblib import load

# Load saved model
loaded_model = load("linear_model.joblib")

# Predict
prediction = loaded_model.predict([[9]])

print("Predicted Trust Score:")
print(prediction[0])
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/multimedia_knowledge_dataset.csv")

# Display original dataset
print("Original Dataset:")
print(df.head())

encoded_df = pd.get_dummies(
    df,
    columns=[
        'text',
        'topic',
        'source_type',
        'sentiment'
    ]
)

# Display encoded dataset
print("\nEncoded Dataset:")
print(encoded_df.head())

encoded_df.to_csv(
    "encoded_multimedia_knowledge_dataset.csv",
    index=False
)

print("\nEncoded dataset saved successfully")
```
