## Data Preprocessing


```python
X = df.drop(['title', 'text', 'date', 'ai_risk_level'], axis=1)
y = df['ai_risk_level']

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Original Target Classes:", label_encoder.classes_)
print("Encoded Target Samples:", y_encoded[:5])

categorical_cols = X.select_dtypes(include='object').columns
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns

for col in categorical_cols:
    X[col] = label_encoder.fit_transform(X[col])

print("\nFeatures after categorical encoding (first 5 rows):")
display(X.head())

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

print("\nShape of training data:", X_train.shape, y_train.shape)
print("Shape of testing data:", X_test.shape, y_test.shape)
```

```python
results = {}
```

## Logistic Regression

```python
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)
log_reg_model.fit(X_train, y_train)
y_pred_log_reg = log_reg_model.predict(X_test)
accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
results["Logistic Regression"] = accuracy_log_reg

print(f"Logistic Regression Accuracy: {accuracy_log_reg:.4f}")
print("Classification Report for Logistic Regression:")
print(classification_report(y_test, y_pred_log_reg, target_names=label_encoder.classes_))
```

## Decision Tree Classifier

```python
dt_classifier_model = DecisionTreeClassifier(random_state=42)
dt_classifier_model.fit(X_train, y_train)
y_pred_dt_classifier = dt_classifier_model.predict(X_test)
accuracy_dt_classifier = accuracy_score(y_test, y_pred_dt_classifier)
results["Decision Tree Classifier"] = accuracy_dt_classifier

print(f"Decision Tree Classifier Accuracy: {accuracy_dt_classifier:.4f}")
print("Classification Report for Decision Tree Classifier:")
print(classification_report(y_test, y_pred_dt_classifier, target_names=label_encoder.classes_))
```

## Support Vector Machine (SVC)

```python
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
results["Support Vector Machine"] = accuracy_svm

print(f"Support Vector Machine Accuracy: {accuracy_svm:.4f}")
print("Classification Report for Support Vector Machine:")
print(classification_report(y_test, y_pred_svm, target_names=label_encoder.classes_))
```

## Naive Bayes

```python
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
accuracy_nb = accuracy_score(y_test, y_pred_nb)
results["Naive Bayes"] = accuracy_nb

print(f"Naive Bayes Accuracy: {accuracy_nb:.4f}")
print("Classification Report for Naive Bayes:")
print(classification_report(y_test, y_pred_nb, target_names=label_encoder.classes_))
```

## K-Nearest Neighbors (KNN)

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
results["K-Nearest Neighbors"] = accuracy_knn

print(f"K-Nearest Neighbors Accuracy: {accuracy_knn:.4f}")
print("Classification Report for K-Nearest Neighbors:")
print(classification_report(y_test, y_pred_knn, target_names=label_encoder.classes_))
```

## K-Means Clustering

K-Means is an unsupervised clustering algorithm, meaning it groups data points into clusters without using a labeled target variable. It is typically used for exploratory data analysis or as a preprocessing step, not for direct classification comparison with supervised models. Therefore, we will demonstrate its application but will not include it in the accuracy comparison for supervised classification models.

```python
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_labels = kmeans_model.fit_predict(X)

print("K-Means cluster assignments (first 20 samples):")
print(kmeans_labels[:20])
print("\nK-Means inertia (sum of squared distances of samples to their closest cluster center):", kmeans_model.inertia_)
```

## Random Forest Classifier

```python
rf_classifier_model = RandomForestClassifier(random_state=42)
rf_classifier_model.fit(X_train, y_train)
y_pred_rf_classifier = rf_classifier_model.predict(X_test)
accuracy_rf_classifier = accuracy_score(y_test, y_pred_rf_classifier)
results["Random Forest Classifier"] = accuracy_rf_classifier

print(f"Random Forest Classifier Accuracy: {accuracy_rf_classifier:.4f}")
print("Classification Report for Random Forest Classifier:")
print(classification_report(y_test, y_pred_rf_classifier, target_names=label_encoder.classes_))
```

## Model Comparison and Best Model Selection

```python
print("\n--- Model Comparison (Accuracy) ---")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.4f}")

best_model_name = max(results, key=results.get)
best_accuracy = results[best_model_name]

print(f"\nThe best performing classification model is {best_model_name} with an accuracy of {best_accuracy:.4f}.")

if best_model_name == "Logistic Regression":
    best_model = log_reg_model
elif best_model_name == "Decision Tree Classifier":
    best_model = dt_classifier_model
elif best_model_name == "Support Vector Machine":
    best_model = svm_model
elif best_model_name == "Naive Bayes":
    best_model = nb_model
elif best_model_name == "K-Nearest Neighbors":
    best_model = knn_model
elif best_model_name == "Random Forest Classifier":
    best_model = rf_classifier_model
else:
    best_model = None

if best_model:
    pickle_filename = f"{best_model_name.replace(' ', '_').lower()}_model.pkl"
    with open(pickle_filename, 'wb') as file:
        pickle.dump(best_model, file)
    print(f"\nBest model '{best_model_name}' saved as '{pickle_filename}'")
else:
    print("No best model found to save.")
```
