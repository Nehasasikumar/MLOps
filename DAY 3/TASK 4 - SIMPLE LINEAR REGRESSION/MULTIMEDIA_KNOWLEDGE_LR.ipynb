```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/multimedia_knowledge_dataset.csv")

# Input and output
X = df[['source_credibility']]
y = df['trust_score']

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Prediction
prediction = model.predict([[9]])

print("Predicted Trust Score:")
print(prediction[0])

# Visualization
plt.scatter(df['source_credibility'], df['trust_score'])

plt.plot(df['source_credibility'],
         model.predict(X))

plt.xlabel("Source Credibility")
plt.ylabel("Trust Score")
plt.title("Simple Linear Regression")

plt.show()
```
