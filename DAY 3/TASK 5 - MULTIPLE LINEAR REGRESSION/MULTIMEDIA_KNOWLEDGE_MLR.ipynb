```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/multimedia_knowledge_dataset.csv")

# Multiple input features
X = df[['source_credibility',
        'article_length',
        'contradiction_score',
        'semantic_similarity',
        'readability_score']]

# Output
y = df['trust_score']

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict new value
prediction = model.predict([[8, 3000, 0.4, 0.8, 75]])

print("Predicted Trust Score:")
print(prediction[0])

# Visualization
plt.scatter(y, model.predict(X))

plt.xlabel("Actual Trust Score")
plt.ylabel("Predicted Trust Score")
plt.title("Multiple Linear Regression")

plt.show()
```
